**Train Test Split**

```python
from sklearn import preprocessing

data['encoded_signals'] = data['signal_outcome'] 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
data['encoded_signals']= label_encoder.fit_transform(data['encoded_signals'])
  
data['encoded_signals'].unique()
```

```python

X = data.drop(columns=['signal_outcome','encoded_signals',''], axis=1)
y = data['encoded_signals']
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# fit model no training data

model = XGBClassifier()
model.fit(X_train, y_train)
# make predictions for test data
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_test, predictions)
print(accuracy*100)
```
