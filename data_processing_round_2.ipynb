* **Normalisation of Adj close price**





```python
data1=data
data2 = pd.DataFrame()
# data1['Adj Close']['nAAPL'] =  data1['Adj Close']['AAPL']  / data1['Adj Close']['AAPL'].abs().max()
A =  data1['Adj Close']['AAPL']  / data1['Adj Close']['AAPL'].abs().max()
data2['nAdjCloseAAPL'] = pd.DataFrame(A)
B =  data1['Adj Close']['INTC']  / data1['Adj Close']['INTC'].abs().max()
data2['nAdjCloseINTC'] = pd.DataFrame(B)
A =  data1['Adj Close']['MSFT']  / data1['Adj Close']['MSFT'].abs().max()
data2['nAdjCloseMSFT'] = pd.DataFrame(A)


data2['AdjCloseAAPL'] = data1['Adj Close']['AAPL']
data2['AdjCloseINTC'] = data1['Adj Close']['INTC']
data2['AdjCloseMSFT'] = data1['Adj Close']['MSFT']

print(data2)
# data1['Adj Close']['nINTC'] =  data1['Adj Close']['INTC']  / data1['Adj Close']['INTC'].abs().max()
# data1['Adj Close']['nMSFT'] =  data1['Adj Close']['MSFT']  / data1['Adj Close']['MSFT'].abs().max()
```

```python
data2.sort_values(by='Date')
print(data2)
```

* **Calculating EMA, RSI, MACD values**

```python
# Calculate EMA with a period of 50, 200, 13, 26
ema_values50 = talib.EMA(data2['nAdjCloseAAPL'], timeperiod=50)
ema_values200 = talib.EMA(data2['nAdjCloseAAPL'], timeperiod=200)
ema_values13 = talib.EMA(data2['nAdjCloseAAPL'], timeperiod=13)
ema_values26 = talib.EMA(data2['nAdjCloseAAPL'], timeperiod=26)

# Combine the EMA values with the original dataframe
data2['AAPLema50'] = ema_values50
data2['AAPLema200'] = ema_values200
data2['AAPLema13'] = ema_values13
data2['AAPLema26'] = ema_values26

ema_values12 = talib.EMA(data2['nAdjCloseAAPL'], timeperiod=12)
ema_values9 = talib.EMA(data2['nAdjCloseAAPL'], timeperiod=9)
data2['AAPLema12'] = ema_values12
data2['AAPLema9'] = ema_values9
```

```python
# Calculate EMA with a period of 50, 200, 13, 26
ema_values50 = talib.EMA(data2['nAdjCloseINTC'], timeperiod=50)
ema_values200 = talib.EMA(data2['nAdjCloseINTC'], timeperiod=200)
ema_values13 = talib.EMA(data2['nAdjCloseINTC'], timeperiod=13)
ema_values26 = talib.EMA(data2['nAdjCloseINTC'], timeperiod=26)
# Combine the EMA values with the original dataframe
data2['INTCema50'] = ema_values50
data2['INTCema200'] = ema_values200
data2['INTCema13'] = ema_values13
data2['INTCema26'] = ema_values26

ema_values12 = talib.EMA(data2['nAdjCloseINTC'], timeperiod=12)
ema_values9 = talib.EMA(data2['nAdjCloseINTC'], timeperiod=9)
data2['INTCema12'] = ema_values12
data2['INTCema9'] = ema_values9
```

```python
# Calculate EMA with a period of 50, 200, 13, 26
ema_values50 = talib.EMA(data2['nAdjCloseMSFT'], timeperiod=50)
ema_values200 = talib.EMA(data2['nAdjCloseMSFT'], timeperiod=200)
ema_values13 = talib.EMA(data2['nAdjCloseMSFT'], timeperiod=13)
ema_values26 = talib.EMA(data2['nAdjCloseMSFT'], timeperiod=26)
# Combine the EMA values with the original dataframe
data2['MSFTema50'] = ema_values50
data2['MSFTema200'] = ema_values200
data2['MSFTema13'] = ema_values13
data2['MSFTema26'] = ema_values26

ema_values12 = talib.EMA(data2['nAdjCloseMSFT'], timeperiod=12)
ema_values9 = talib.EMA(data2['nAdjCloseMSFT'], timeperiod=9)
data2['MSFTema12'] = ema_values12
data2['MSFTema9'] = ema_values9
```

```python
#RSI values for period 14
rsi_period1 = 14

rsi = talib.RSI(data2['AdjCloseAAPL'], rsi_period1)
# Add RSI values to dataframe
data2['AAPLrsi14'] = rsi
```

```python
#RSI values for period 14
rsi_period1 = 14

rsi = talib.RSI(data2['AdjCloseINTC'], rsi_period1)
# Add RSI values to dataframe
data2['INTCrsi14'] = rsi
```

```python
#RSI values for period 14
rsi_period1 = 14

rsi = talib.RSI(data2['AdjCloseMSFT'], rsi_period1)
# Add RSI values to dataframe
data2['MSFTrsi14'] = rsi
```

```python
#MACD values

data2['AAPLmacd'] = data2['AAPLema12'] - data2['AAPLema26']
data2['AAPLmacdSignal'] = data2['AAPLema9']

data2['INTCmacd'] = data2['INTCema12'] - data2['INTCema26']
data2['INTCmacdSignal'] = data2['INTCema9']

data2['MSFTmacd'] = data2['MSFTema12'] - data2['MSFTema26']
data2['MSFTmacdSignal'] = data2['MSFTema9']
```

 ## **Section for APPLE backtest, dataset**

*   Backtest data with conditions
*   Create dataset with signals to optimize 



```python
buy_position = 0     # 0 means no position
orders = []
order_index = []
```

```python
# BT for AAPL

# ol
for row in range(0,data2.shape[0]):
  buy_flags = 0
  sell_flags = 0
  adj_close = data2['AdjCloseAAPL'].iloc[row]
  AAPLema13 = data2['AAPLema13'].iloc[row]
  AAPLema26 = data2['AAPLema26'].iloc[row]
  AAPLema50 = data2['AAPLema50'].iloc[row]
  AAPLema200 = data2['AAPLema200'].iloc[row]
  AAPLrsi14 = data2['AAPLrsi14'].iloc[row]
  AAPLmacd = data2['AAPLmacd'].iloc[row]
 
  # buy conditions
  if (AAPLrsi14 > 30 and AAPLrsi14 < 60):
    print("row no ", row, " rsi 30 to 60 ", " buy flags ", buy_flags )
    buy_flags = buy_flags + 1
  
  if (AAPLema50 > AAPLema200):
    print("row no ", row, " 50 > 200 ", " buy flags ", buy_flags )
    buy_flags = buy_flags + 1

  if (AAPLema13 > AAPLema26):
    print("row no ", row, " 13 > 26 ", " buy flags ", buy_flags )
    buy_flags = buy_flags + 1

  if (AAPLmacd > 0):
    print("row no ", row, " macd > 0 ", " buy flags ", buy_flags )
    buy_flags = buy_flags + 1
  
  #sell conditions
  if (AAPLrsi14 > 60):
    print("row no ", row, " rsi > 60 ", " sell flags ", sell_flags )
    sell_flags = sell_flags + 1

  if (AAPLema50 < AAPLema200):
    print("row no ", row, " 50 < 200 ", " sell flags ", sell_flags, )
    sell_flags = sell_flags + 1

  if (AAPLema13 < AAPLema26):
    print("row no ", row, " 13 < 26 ", " sell flags ", sell_flags, )
    sell_flags = sell_flags + 1

  if (AAPLmacd < 0):
    print("row no ", row, " macd < 0 ", " sell flags ", sell_flags, )
    sell_flags = sell_flags + 1

  # print("Row no ", row, "AAPLema13 ", AAPLema13, "AAPLema26 ", AAPLema26, "AAPLema50 ", AAPLema50, "AAPLema200 ", AAPLema200, "AAPLrsi14 ", AAPLrsi14, "AAPLmacd ", AAPLmacd, "buy flags ", buy_flags, "Sell flags ", sell_flags)

  if buy_position == 0 and buy_flags >= 3:
    order_index = [row]
    buy_position = 1
    buy_flags = 0
    sel_flags = 0

  elif buy_position > 0 and sell_flags >= 3:
    order_index.append(row)
    orders.append(order_index)
    buy_flags = 0
    buy_position = 0
    sell_flags = 0


print(orders)
```

```python
print(len(orders))
```

```python
count = 0
totalPnl = 0 

#structure of i after loop becomes like : entry_index 0 , exit_index 1, entry_price 2, exit_price 3 , pnl 4
for i in orders:
  count = count + 1
  print("Order - ", count)
  print("Row no ", i[0], " Adj Close entry ", data2['AdjCloseAAPL'].iloc[i[0]], "AAPLema13 ", data2['AAPLema13'].iloc[i[0]], "AAPLema26 ", data2['AAPLema26'].iloc[i[0]], "AAPLema50 ", data2['AAPLema50'].iloc[i[0]], "AAPLema200 ", data2['AAPLema200'].iloc[i[0]], "AAPLrsi14 ", data2['AAPLrsi14'].iloc[i[0]] , "AAPLmacd ", data2['AAPLmacd'].iloc[i[0]])
  print("Row no ", i[1], " Adj Close entry ", data2['AdjCloseAAPL'].iloc[i[1]], "AAPLema13 ", data2['AAPLema13'].iloc[i[1]], "AAPLema26 ", data2['AAPLema26'].iloc[i[1]], "AAPLema50 ", data2['AAPLema50'].iloc[i[1]], "AAPLema200 ", data2['AAPLema200'].iloc[i[1]], "AAPLrsi14 ", data2['AAPLrsi14'].iloc[i[1]] , "AAPLmacd ", data2['AAPLmacd'].iloc[i[1]])
  #print('================PNL =',data2['AdjCloseAAPL'].iloc[i[1]] - data2['AdjCloseAAPL'].iloc[i[0]] ,' ================')
  pnl = data2['AdjCloseAAPL'].iloc[i[1]] - data2['AdjCloseAAPL'].iloc[i[0]]
  i.append(data2['AdjCloseAAPL'].iloc[i[0]])
  i.append(data2['AdjCloseAAPL'].iloc[i[1]])
  i.append(pnl)
  print('=================current pnl ', i[4] , ' ================')
  totalPnl = totalPnl + pnl

print(totalPnl)
```

```python
data2['orders_executed'] = 0
data2['pnl_magnitude'] = 0
data2['signal_outcome'] = 0

for row in range(data2.shape[0]):
  data2.loc[row, 'orders_executed'] = 0
  data2.loc[row, 'pnl_magnitude'] = 0
  data2.loc[row, 'signal_outcome'] = 0

for i in orders:
  buy_index = i[0]
  sell_index = i[1]

  data2['orders_executed'].iloc[buy_index] = 1
  data2['orders_executed'].iloc[sell_index] = -1

  data2['pnl_magnitude'].iloc[buy_index] = i[4] 

  if i[4] < 0 :
    data2['signal_outcome'].iloc[buy_index] = -1
  
  else :
    data2['signal_outcome'].iloc[buy_index] = 1

print(data2.iloc[175], data2.iloc[191])
```
